```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("../dataset/Mall_Customers.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker='o')

plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)

y_kmeans = kmeans.fit_predict(X)
```

```python
df['Cluster'] = y_kmeans
```

```python
plt.figure(figsize=(10,6))

plt.scatter(X.iloc[:,0], X.iloc[:,1], c=y_kmeans)

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

plt.title("Customer Segments")

plt.show()
```

```python
print(kmeans.cluster_centers_)
```

```python
# Insights

1. Customers were divided into 5 segments.

2. Some customers have high income and high spending score.

3. Some customers have high income but low spending score.

4. Businesses can target each customer group differently.
```
